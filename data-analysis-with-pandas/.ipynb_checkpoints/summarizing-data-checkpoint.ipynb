### Summarizing Major Categories
 __Description__<br>
Let's start by understanding the number of people in each Major_category for both data sets.
     
__Instructions__<br>
- Return the unique values in Major_category.
- For each unique value:
    - Return all of the rows where Major_category equals that unique value.
    - Calculate the total number of students those rows represent (using the Total column).
    - Keep track of the totals by adding the Major_category value and the total number of students to a dictionary.

```python
aa_cat_counts = {}
rg_cat_counts = {}

def calc_major_category_total(dataset):
    major_category = dataset["Major_category"]
    unique_categories = major_category.unique()
    counts_dictionary = {}
    
    
    for i in unique_categories:
        # Return all of the rows where Major_category equals that unique value.
        major_category_equals_unique_value = dataset[major_category == i]
        # Calculate the total number of students those rows represent (using the Total column).
        total = major_category_equals_unique_value["Total"].sum()
        # Keep track of the totals by adding the Major_category value and the total number of students to a dictionary.
        counts_dictionary[i] = total
    return counts_dictionary
        
aa_cat_counts = calc_major_category_total(all_ages)
rg_cat_counts = calc_major_category_total(recent_grads)
```

###  Low-Wage Job Rates
    Instructions

- Use the Low_wage_jobs and Total columns to calculate the proportion of recent college graduates that worked low wage jobs.

```python
total_sum = recent_grads["Total"].sum()
low_wage_jobs_sum = recent_grads["Low_wage_jobs"].sum()

low_wage_proportion = low_wage_jobs_sum / total_sum
print(low_wage_proportion)
```

###  Comparing Data Sets
Next, let's calculate the number of majors where recent graduates did better than the overall population.
    
    Instructions


    Use a for loop to iterate over majors.
        For each major, use Boolean filtering to find the corresponding row in both DataFrames.
        Compare the values for Unemployment_rate to see which DataFrame has a lower value.
        Increment rg_lower_count if the value for Unemployment_rate is lower for recent_grads than it is for all_ages.

    Display rg_lower_count with the print() function.


```python
rg_lower_count = 0
majors = recent_grads["Major"].unique()

for i in majors:
    
    rg_row_major = recent_grads[majors == i]
    aa_row_major = all_ages[all_ages["Major"] == i]
    
    
    rg_unemployement_rate = rg_row_major["Unemployment_rate"].sum()
    aa_unemployement_rate = aa_row_major["Unemployment_rate"].sum()
    
    if rg_unemployement_rate < aa_unemployement_rate:
        rg_lower_count += 1
print(rg_lower_count)
```

__Result__ :
It appears that less recent graduates who studied 44 of the 173 majors ended up having lower unemployment rates than the general population.
