### Apply() Logic Over the Columns in a Dataframe
    Instructions
In the following code, we use a lambda function to multiply each float column by 2:

```python
types = fandango_films.dtypes
float_types = types[types.values == 'float64'].index
float_df = fandango_films[float_types]
col_x_two = float_df.apply(lambda x: x*2)
deviations =float_df.apply(lambda x: np.std(x))
```
