### Formatting And Spacing
To tweak the dimensions of the plotting area, we need to use the figsize parameter when we call plt.figure():<br>
This parameter takes in a tuple of floats
 
    Intructions
- For the plot we generated in the last screen, set the width of the plotting area to 12 inches and the height to 5 inches.

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2, 1,2)
ax1.plot(unrate["DATE"][0:12], unrate["VALUE"][0:12])
ax2.plot(unrate["DATE"][12:24],unrate["VALUE"][12:24])
```

###  Comparing Across More Years
    Instructions
- Generate a grid with 5 rows and 1 column and plot data from the individual years. Start with 1948 in the top subplot and end with 1952 in the bottom subplot.

```python
fig = plt.figure(figsize=(15, 12))
ax1 = fig.add_subplot(5,1,1)
ax2 = fig.add_subplot(5,1,2)
ax3 = fig.add_subplot(5,1,3)
ax4 = fig.add_subplot(5,1,4)
ax5 = fig.add_subplot(5,1,5)
ax1.plot(unrate["DATE"][0:12], unrate["VALUE"][0:12])
ax2.plot(unrate["DATE"][12:24],unrate["VALUE"][12:24])
ax3.plot(unrate["DATE"][24:36],unrate["VALUE"][24:36])
ax4.plot(unrate["DATE"][36:48],unrate["VALUE"][36:48])
ax5.plot(unrate["DATE"][48:60],unrate["VALUE"][48:60])
```

### Overlaying Line Charts
    Instructions
- Generate line charts in the base subplot, using the MONTH column for the x-axis instead of the DATE column

```python
unrate["MONTH"] = unrate["DATE"].dt.month
```

```python
fig = plt.figure(figsize=(15,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start = (i*12)
    end = (i+1)*12
    subset = unrate[start:end]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i])
```

### Adding a Legend
    Instructions
- Modify the code from the last screen that overlaid 5 plots to include a legend. Use the year value for each line chart as the label

```python
fig = plt.figure(figsize=(15,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start = (i*12)
    end = (i+1)*12
    subset = unrate[start:end]
    #  the year value for each line chart as the label
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
plt.legend(loc = 'upper left')
```

### Final Tweaks
    Instructions
- Modify the code from the last screen:

    - Set the title to "Monthly Unemployment Trends, 1948-1952".
    - Set the x-axis label to "Month, Integer".
    - Set the y-axis label to "Unemployment Rate, Percent".

```python
fig = plt.figure(figsize=(15,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start = (i*12)
    end = (i + 1)*12
    subset = unrate[start:end]
    label = str(1948 + i)
    plt.plot(subset["MONTH"], subset["VALUE"],c = colors[i], label= label)
plt.legend(loc = 'upper left')
plt.title(s='Monthly Unemployment Trends, 1948-1952')
plt.xlabel(s='Month, Integer')
plt.ylabel(s='Unemployment Rate, Percent')
```
