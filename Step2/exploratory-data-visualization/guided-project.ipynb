# Visualizing Earnings Based On College Majors
Learn how to use pandas to quickly create visualizations.

```python
# importing the libraries we need
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline
```

```python
# read the dataset
df = pd.read_csv('/home/aida/Desktop/Dataquest/Step2/exploratory-data-visualization/data/recent-grads.csv')
#df.head(); df.tail(); df.describe();df.iloc[0];
```

```python
# drop rows with missing values
df = df.dropna()
df.shape
```

```python
df.plot(x="Sample_size", y="Median", kind='scatter', title = 'Sample size vs Median', figsize=(10,5))
```

```python
df.plot(x="Sample_size", y="Unemployment_rate", kind='scatter', title = 'Sample size vs Unemployment_rate', figsize=(10,5))
```

```python
df.plot(x="Full_time", y="Median", kind='scatter', title = 'Full time vs Median', figsize=(10,5))
```

```python
df.plot(x="ShareWomen", y="Unemployment_rate", kind='scatter', title = 'ShareWomen vs Unemployment_rate', figsize=(10,5))
```

```python
df.plot(x="Men", y="Median", kind='scatter', title = 'Men vs Median', figsize=(10,5))
```

```python
df.plot(x="Women", y="Median", kind='scatter', title = 'Women vs Median', figsize=(10,5))
```

#### Do students in more popular majors make more money?

```python
columns = [
    "Sample_size",
    "Median",
    "Employed",
    "Full_time",
    "ShareWomen",
    "Unemployment_rate",
    "Men",
    "Women"]
fig = plt.figure(figsize=(5,12))
for i in range(1,5):
    ax = fig.add_subplot(4,1,i)
    ax = df[columns[i]].plot(kind="hist")
```

```python
fig = plt.figure(figsize=(5,12))
for i in range(4,8):
    ax = fig.add_subplot(4,1,i-3)
    ax = df[columns[i]].plot(kind="hist")
```

### Pandas, Scatter Matrix Plot
- Create a 2 by 2 scatter matrix plot using the Sample_size and Median columns.
- Create a 3 by 3 scatter matrix plot using the Sample_size, Median, and Unemployment_rate columns.

```python
scatter_matrix(df[['Sample_size','Median']], figsize = (6,6), alpha = 0.2)
```

```python
scatter_matrix(df[['Sample_size', 'Median', 'Unemployment_rate']], diagonal='kde')
```
