### Figuring Out What Pies People Eat
    Instructions  
- We can find out how many people eat one of these three pies for Thanksgiving dinner by figuring out for how many people all three columns are null.


```python
apple = data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple"]
pumpkin = data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin"]
pecan = data["Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan"]
apple_isnull = apple.isnull()
pumpkin_isnull = pumpkin.isnull()
pecan_isnull = pecan.isnull()
no_pies = apple_isnull & pumpkin_isnull & pecan_isnull
no_pies.value_counts()
```

```python
# Where that Series contains False, the person ate at least one of the types of pies for Thanksgiving dinner. Where it contains True, they ate none of the types of pies.
ate_pie = pd.isnull(apple) & pd.isnull(pumpkin) & pd.isnull(pecan)
ate_pie.value_counts()
```

###  Converting Age To Numeric
    Instructions
- Write a function to convert a single string to an appropriate integer value. This will allow us to convert the values in the Age column to integers.

```python
age = data["Age"]
age.value_counts()
```

```python
def string_to_int(age_string):
    if pd.isnull(age_string):
        return None
    age_string = age_string.split(" ")[0]
    age_string = age_string.replace("+", "")
    return int(age_string)

data["int_age"] = age.apply(string_to_int)
data["int_age"].describe()
```

### Converting Income To Numeric
- The How much total combined money did all members of your HOUSEHOLD earn last year? column is very similar to the Age column. It contains categories, but can be converted to numerical values. 

```python
income = data["How much total combined money did all members of your HOUSEHOLD earn last year?"]
income.value_counts()
```

```python
def extract_string(str_income):
    if pd.isnull(str_income):
        return None
    str_income = str_income.split(" ")[0]
    if str_income == "Prefer":
        return None
    str_income = str_income.replace("$", "")
    str_income = str_income.replace(",", "")
    return int(str_income)

data["int_income"] = income.apply(extract_string)   
data["int_income"].describe()    
```

### Correlating Travel Distance And Income
